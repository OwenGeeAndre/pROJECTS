# Multiple Linear Regession

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

# Loading the Data

```python
data = pd.read_csv(r'/content/1.02.+Multiple+linear+regression.csv')
```

```python
data.head(3)
```

# Create the Multiple Linear Regression

```python
x = data.drop(columns="GPA")
y = data["GPA"]
```

## Training the Model

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

## Calculating the $R^2$ and Adjusted $R^2$

```python
reg.score(x,y)
```

### Formula for Adjusted R-squared

$R^2_{adj.} = 1 - (1-R^2)*\frac{n-1}{n-p-1}$

> n = 84 (number of observations)
>
> p = 2 (number of predictors)

```python
r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]
```

```python
adj_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adj_r2
```

## Feature Selection: F_regression

>F-regression creates simple linear regressions
of each feature and the dependent variable

```python
from sklearn.feature_selection import f_regression
```

```python
f_values, p_values = f_regression(x,y)
p_values
```

```python
p_values.round(3)
```

* The $p_{value}$ of `SAT`is 0 and `Rand 1,2,3` is 0.676
* with $p_{value} > 0.05$, `Rand 1,2,3` is not a useful predictor

# Creating a summary table

```python
reg_summary = pd.DataFrame(data=x.columns.values, columns=["Features"])
reg_summary
```

```python
reg_summary["coefficients"] = reg.coef_
reg_summary["p-values"] = p_values.round(3)
reg_summary
```
